```python
# from geopy.geocoders import Nominatim
import googlemaps

# Todo: replace with your API key
geolocator = googlemaps.Client(key="your_api_key_here")

def get_location(place_name,province):
    """
    Get latitude and longitude for a given place name using Google Maps API.
    
    Args:
        place_name (str): Name of the place
        
    Returns:
        tuple: (latitude, longitude, url) or (None, None, None) if not found
    """
    try:
        locations = geolocator.geocode(f"{place_name}, โรงพยาบาล, {province}, Thailand")
        if locations:
            location = locations[0]
            lat = location['geometry']['location']['lat']
            lng = location['geometry']['location']['lng']
            url = f"https://www.google.com/maps/search/?api=1&query={lat},{lng}&query_place_id={location['place_id']}"
            return (lat, lng, url)
        else:
            return (None, None, None)
    except:
        return (None, None, None)

# Initialize columns if they don't exist
if 'latitude' not in df.columns:
    df['latitude'] = None
if 'longitude' not in df.columns:
    df['longitude'] = None
if 'google_maps_url' not in df.columns:
    df['google_maps_url'] = None

# Loop through rows and get location only if latitude or longitude is empty
for idx, row in df.iterrows():
    if pd.isna(row['latitude']) or pd.isna(row['longitude']):
        lat, lng, url = get_location(row['โรงพยาบาล'], row['จังหวัด'])
        df.at[idx, 'latitude'] = lat
        df.at[idx, 'longitude'] = lng
        df.at[idx, 'google_maps_url'] = url

df.head()

df.to_csv('sso 2026.csv', index=False)
```
